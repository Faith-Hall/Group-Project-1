```python
# Import dependencies 
import pandas as pd
from pathlib import Path
```

```python
# reference file path
Energy_path = Path("Resources/owid-energy-data.csv")
# import as df
energy_df = pd.read_csv(Energy_path)
energy_df.head()
```

```python
LE = Path("Resources/WPP2022_Demographic_Indicators_Medium.csv")
LE_df = pd.read_csv(LE)
LE_df.head()
```

```python
LE_columns = [col for col in LE_df.columns if col.startswith("LE") and "Male" not in col and "Female" not in col]
LE_columns
```

```python
Meta_columns = ['SortOrder', 'Location', 'Time']
Meta_columns
```

```python
# Concatenate meta columns and life expectancy columns
LE_df = LE_df[Meta_columns + LE_columns]
LE_df.head()
```

```python
# ?
LE_df["Location"].value_counts()
```

```python
# Renaming columns
LE_df.rename(columns={"SortOrder": "Record Counter","Time": "Year"},inplace=True)
LE_df.head()
```

```python
# Filter Demo years to 2022 and earlier
LE_df[(LE_df['Year'] <= 2022)]
LE_df.head()
```

```python
# Filter energy df years
energy_df[(energy_df['year'] >= 1950)]
energy_df.head()
```

```python
energy_meta_columns = ['country', 'year', 'population', 'fossil_energy_per_capita', 'renewables_energy_per_capita', 'energy_per_capita']

energy_df_clean = energy_df[energy_meta_columns]
energy_df_clean.head()
```

```python
# Rename columns
energy_df_clean.rename(columns={"country": "Location","year": "Year"},inplace=True)
energy_df_clean.head()
```

```python
merge_df = pd.merge(energy_df_clean, LE_df)
merge_df.head()
```

```python
# create csv
merge_df.to_csv('energyLE.csv')
```

```python
# drop NaN 
merge_df_clean = merge_df.dropna()
merge_df_clean.head()
```

```python
merge_df_clean.to_csv('energyLEclean.csv')
```

```python
# Rename columns 
merge_df_clean.rename(columns={"population": "Population",
                               "fossil_energy_per_capita": "Fossil-energy-per-capita", 
                               "renewables_energy_per_capita": "Renewables-energy-per-capita",
                               "energy_per_capita": "Energy-per-capita",
                               "LEx": "Life-expectancy-at-birth",
                               "LE15": "Life-expetancy-age-15",
                               "LE65": "Life-expetancy-age-65",
                               "LE80": "Life-expetancy-age-80"},inplace=True)
merge_df_clean.head()
```
