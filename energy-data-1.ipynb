```python
# Import dependencies 
import pandas as pd
from pathlib import Path
```

```python
# reference file path
Energy_path = Path("Resources/owid-energy-data.csv")
# import as df
energy_df = pd.read_csv(Energy_path)
energy_df.head()
```

```python
Demo_path = Path("Resources/WPP2022_Demographic_Indicators_Medium.csv")
Demo_df = pd.read_csv(Demo_path)
```

```python
Demo_df.columns

LE_columns = [col for col in Demo_df.columns if col.startswith("LE") and "Male" not in col and "Female" not in col]
LE_columns
```

```python
Meta_columns = ['SortOrder', 'Location', 'Time']
Meta_columns
```

```python
# Concatenate meta columns and life expectancy columns
Demo_df = Demo_df[Meta_columns + LE_columns]
Demo_df.head()
```
